```python
import matplotlib.pyplot as plt

# Marks scored by the student in two exams
E1 = [20, 18, 30, 22, 15]
E2 = [19, 22, 33, 30, 19]

# Calculate the difference in marks between the two exams for each student
difference = [E2[i] - E1[i] for i in range(len(E1))]

# Names of the students
students = ['Student 1', 'Student 2', 'Student 3', 'Student 4', 'Student 5']

# Plotting the bar chart
plt.bar(students, difference, color='green')
plt.xlabel('Students')
plt.ylabel('Difference in Marks')
plt.title('Difference in Marks Between Two Exams for Each Student')
plt.show()
```
